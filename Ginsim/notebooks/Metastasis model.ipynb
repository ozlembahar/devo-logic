### 1. Visualize results on a PCA

````python
## Use biolqm to visualize the stable states
model_biolqm = ginsim.to_biolqm(model_ginsim) # convert to biolqm

## Compute fixpoints
fps = biolqm.fixpoints(model_biolqm)
print(len(fps), "fixpoints") # shows the number of fixpoints
pd.DataFrame(fps)
#tabulate(fps) # table displaying all the fixpoints
````

Use a PCA representation to visualize how close the stable states are 

````python
# Getting the list of nodes
nodes = list(fps[0].keys())

# Building a 2D numpy array with the nodes order defined above, with each fixpoint value
data = np.array([[fp[node] for node in nodes] for fp in fps])

# The array is XX fixpoints * XX nodes
data.shape

from sklearn import decomposition
from sklearn.decomposition import PCA
#pca = PCA(n_components=3)
#pca_out = pca.fit(data)

from pandas import DataFrame

# training the PCA
pca = decomposition.PCA()
pca.fit(data)
````

````python
## Visualise the results in the form of a graph
# Getting the new coordinates
X_pca = pca.transform(data)

# Plotting the steady states
plt.figure(figsize=(3, 4), dpi=200)
plt.plot(X_pca[:, 0], X_pca[:, 1], ".")

# Printing arrows, but there is too much. We need to select some nodes
#some_nodes = ["EGFR", "TP53","CDK1"]
some_nodes = ["Metastasis", "EMT","Apoptosis","CDH1"]
for some_node in some_nodes:

    v = pca.components_[0:2, nodes.index(some_node)].T
    plt.arrow(0, 0, v[0], v[1], color='red', width=0.005)
    plt.text(v[0], v[1], some_node, color='black', ha='center', va='center', fontsize=5)

    
plt.xlabel("PC1: %.2f%% variance explained" % (pca.explained_variance_ratio_[0]))
plt.ylabel("PC2: %.2f%% variance explained" % (pca.explained_variance_ratio_[1]))
````

The PCA of the stable states shows that there are at least 3 clusters defined by an apoptotic phenotype (3 on the left), 2 for EMT phenotype (top right) and 2 for the metastasis phentoype (bottom left). The Homeostatic State (HS) phenotype is not showing here (to improve). 

## MaBoSS simulation of the wild type model

Run MaBoSS now. The files used are:
*model.bnd* and *model.cfg*

````python
model_maboss = biolqm.to_maboss(model_biolqm) # convert to maboss
model_maboss.update_parameters(sample_count=10000,max_time=100, time_tick=0.1)
````

Load the model with either all 0s or random initial conditions and modify some configuration parameters. Note that if you used the export of GINsim export, all the initial values will be set to 0

````python
## Load the Wild-Type model
#WT = maboss.load(bnd_file, cfg_file)

## modify the max_time if necessary
## select the variables considered to be internal (not appearing in the resulting csv file) 
## run the model for random initial conditions (defined in the cfg file) or from all 0s initial conditions 

WT_allinputs = model_maboss.copy()
````

### Set initial conditions reported in the article

We set the miRNA at 1 and we let the inputs random. The rest remains at 0 initially.

````python
## Wild Type initial conditions of Cohen et al. 
maboss.set_nodes_istate(WT_allinputs, ["miR200", "miR203","miR34"], [0,1])
maboss.set_nodes_istate(WT_allinputs, ["ECMicroenv", "DNAdamage","GF","TGFbeta"], [0.5,0.5])
````

````python
## Select outputs for visualisation of the results
WT_allinputs.network.set_output(('Metastasis','EMT','Apoptosis','Invasion'))
````

````python
## Run a MaBoSS simulation
run_WT_allinputs = WT_allinputs.run()

## Plot the results of the WT simulation
run_WT_allinputs.plot_trajectory(legend=True)
plt.title('WT model state trajectories')

run_WT_allinputs.get_nodes_probtraj().plot(legend=True)
plt.title('WT node trajectories')
## Note that the model state <nil> means that none of the external variables are ON. It corresponds here to the phenotype Homeostatic State (HS)

# Plot pie chart
run_WT_allinputs.plot_piechart()
plt.title('WT pie chart')

## You may want to save the figure
# figure = run_WT_allinputs.get_states_probtraj().plot()
# save_figure(figure, 'WT')
````

## 4. Predicting genetic interactions

````python
import pypint
from itertools import combinations
from functools import reduce
````

````python
# Libraries for creating combinations of mutants
from itertools import combinations
from functools import reduce

m = biolqm.to_pint(model_biolqm)
m.initial_state["ECMicroenv"] = 1
m.initial_state["DNAdamage"] = {0,1}

mutants = m.oneshot_mutations_for_cut("Apoptosis=1", exclude={"ECMicroenv", "DNAdamage"})
mutants
````

````python
mutant_combinations = [combinations(m.items(), 2) for m in mutants if len(m) >= 2]
candidates = reduce(set.union, mutant_combinations, set())
candidates
````

````python
for mutant in sorted(candidates):
    mut_sim = WT_allinputs.copy()
    for (node,value) in mutant:
        mut_sim.mutate(node, "ON" if value else "OFF")
    mut_res = mut_sim.run()
    mut_res.plot_piechart(embed_labels=False, autopct=4)
    mutant_name = "/".join(["%s:%s"%m for m in mutant])
    plt.title("%s" % mutant_name)
````
